### Deliveries Analysis

```python
deliveries
```

```python
deliveries.dtypes
```

```python
# Make sure 'over' is float
deliveries['over'] = deliveries['over'].astype(float)
```

```python
mask6 = deliveries['ball'] == 6
mask_other = deliveries['ball'] != 6

deliveries.loc[mask6, 'over'] += 1
deliveries.loc[mask_other, 'over'] += deliveries.loc[mask_other, 'ball'] / 10
```

```python
# Grouping by match_id with scores
M = ['match_id','batsman_runs', 'extra_runs', 'total_runs', 'is_wicket']
inning1 = deliveries[deliveries['inning'] == 1]
inning2 = deliveries[deliveries['inning'] == 2]
inn1 = inning1[M].groupby('match_id').sum()
inn2 = inning2[M].groupby('match_id').sum()
i=1
for inn in [inn1, inn2]:
    new_cols = []
    for j in range(len(inn.columns)):
        new_cols.append(inn.columns[j] + f'_{i}')
    inn.columns = new_cols
    i+=1
t1 = inning1[['match_id', 'batting_team']].groupby('match_id').agg('first')
t2 = inning2[['match_id', 'batting_team']].groupby('match_id').agg('first')
i=1
for inn in [t1, t2]:
    new_cols = []
    for j in range(len(inn.columns)):
        new_cols.append(inn.columns[j] + f'_{i}')
    inn.columns = new_cols
    i+=1
o1 = inning1[['match_id', 'over']].groupby('match_id').agg('last')
o2 = inning2[['match_id', 'over']].groupby('match_id').agg('last')
i=1
for inn in [o1, o2]:
    new_cols = []
    for j in range(len(inn.columns)):
        new_cols.append(inn.columns[j] + f'_{i}')
    inn.columns = new_cols
    i+=1
df_matches = pd.concat([t1, inn1, o1, t2, inn2, o2], axis=1)
df_matches['match_id'] = df_matches.index
df_matches.reset_index(drop=True, inplace=True)
df_matches
```

```python
cols = ['match_id', 'batting_team_1', 'batsman_runs_1', 'extra_runs_1', 'total_runs_1',
       'is_wicket_1', 'over_1', 'batting_team_2', 'batsman_runs_2',
       'extra_runs_2', 'total_runs_2', 'is_wicket_2', 'over_2']
df_matches = df_matches[cols]
mat_id = df_matches.match_id.unique()
rep_id = np.arange(1, 1096, 1)
# Create a mapping dictionary from list_1 to list_2
mapping_dict = dict(zip(mat_id, rep_id))
# Replace across multiple columns
df_matches.replace(mapping_dict, inplace=True)
df_matches
```

```python
df_matches.to_csv('./data/match_info.csv', index=False)
```

#### Run rate + wickets bar graph, run graph

```python
ip = 1
match_df = deliveries[deliveries['match_id'] == ip]
inn1 = match_df[match_df['inning'] == 1]
inn2 = match_df[match_df['inning'] == 2]
match_df
```

```python
runs1 = inn1[['over','total_runs', 'is_wicket']]
runs2 = inn2[['over','total_runs', 'is_wicket']]
#print(runs)

# Define the range intervals
start = 0
end = 20
step = 1  # Define the step size for the ranges (0 to 1, 1 to 2, etc.)

# Create bins based on the defined ranges
bins = np.arange(start, end+step, step)  # Bin edges

# Initialize a list to hold the sum of values in each bin for both data sets
bin_sums1 = []
bin_sums2 = []
bin_wkts_sums1 = []
bin_wkts_sums2 = []

# For each bin, sum the values that fall within it for both data sets
for i in range(len(bins) - 1):
    bin_values1 = runs1[(runs1.over >= bins[i]) & (runs1.over < bins[i+1])]
    bin_values2 = runs2[(runs2.over >= bins[i]) & (runs2.over < bins[i+1])]
    
    if len(bin_values1) != 0:
        bin_sums1.append(bin_values1['total_runs'].sum())  # Sum of values for data1 in the bin
        bin_wkts_sums1.append(bin_values1['is_wicket'].sum())

    if len(bin_values2) != 0:
        bin_sums2.append(bin_values2['total_runs'].sum())  # Sum of values for data2 in the bin
        bin_wkts_sums2.append(bin_values2['is_wicket'].sum())    
# bin_sums2
```

```python
# Over by over - 2 innings
fig1 = plt.figure(figsize=(16, 10))

width = 0.35  # Width of each bar

# Create two sets of bars: one for each data set
bar1 = plt.bar(np.arange(1, len(bin_sums1) + 1) - width/2, bin_sums1, width=width, color='skyblue', edgecolor='black', label='1st Inning')
bar2 = plt.bar(np.arange(1, len(bin_sums2) + 1) + width/2, bin_sums2, width=width, color='lightgreen', edgecolor='black', label='2nd Inning')

# Set the labels for x-axis to be 1, 2, 3, ..., matching the bin number
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.title('Runs per over')

# Set x-ticks to correspond to bin numbers (1, 2, 3,...)
plt.xticks(np.arange(1, 21))

m = len(bin_sums1)
n = len(bin_sums2)
# Annotate each bar with the sum of values in that bin
for i in range(20):
    if i < m:
        plt.text(bar1[i].get_x() + bar1[i].get_width() / 2, bin_sums1[i] - 1, str(bin_sums1[i]),
                ha='center', va='bottom', fontsize=10, color='black')
        plt.text(bar1[i].get_x() + bar1[i].get_width() / 2, bin_sums1[i] + 0.5, bin_wkts_sums1[i],
                ha='center', va='bottom', fontsize=10, color='black')
    if i < n:
        plt.text(bar2[i].get_x() + bar2[i].get_width() / 2, bin_sums2[i] - 1, bin_sums2[i],
                ha='center', va='bottom', fontsize=10, color='black')
        plt.text(bar2[i].get_x() + bar2[i].get_width() / 2, bin_sums2[i] + 0.5, bin_wkts_sums2[i],
                ha='center', va='bottom', fontsize=10, color='black')

# Add a legend
plt.legend()

# Show the grid
plt.grid(True)
#plt.show()
```

```python
# Create the line graph with two innings
plt.figure(figsize=(10, 6))

# Plot the first data set
plt.plot(np.arange(1, len(bin_sums1) + 1), bin_sums1, marker='o', linestyle='-', color='green', markerfacecolor='red', label='1st Inning')

# Plot the second data set
plt.plot(np.arange(1, len(bin_sums2) + 1), bin_sums2, marker='s', linestyle='-', color='blue', markerfacecolor='yellow', label='2nd Inning')

# Set the labels and title
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.title('Runs per over')

# Set x-ticks to correspond to bin numbers (1, 2, 3,...)
plt.xticks(np.arange(1, 21))

# Add a legend to differentiate between the two lines
plt.legend()

# Show the grid
plt.grid(True)
plt.show()
```

```python
# Calculate cumulative sum using numpy
cum_sum1 = np.cumsum(bin_sums1)
cum_sum2 = np.cumsum(bin_sums2)
cum_sum2.shape[0]
```

```python
# Run graph
plt.figure()
plt.plot(np.arange(1, cum_sum1.shape[0]+1), cum_sum1, marker='o', linestyle='-', color='green', markerfacecolor='red', label='1st Inning')
plt.plot(np.arange(1, cum_sum2.shape[0]+1), cum_sum2, marker='o', linestyle='-', color='blue', markerfacecolor='yellow', label='2nd Inning')
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.title('Runs graph')
plt.xticks(np.arange(1, 21))
plt.legend(title='Teams', loc='upper left')  # loc can be changed to control position
plt.show()
```
