```python
dp = pd.read_csv('./data/player_data.csv')  # This dataset is better than players1.csv
dp
```

```python
dp.isna().sum()
```

```python
# Converting 'no stats' to 0
dp = dp.replace('No stats', 0)
```

```python
# Matches_Batted & Matches_Bowled are same -> lets drop one and change col name to matches
dp.drop('Matches_Bowled', axis=1, inplace=True)
dp.columns
```

```python
dp.columns = ['Year', 'Player_Name', 'Matches', 'Not_Outs', 'Runs_Scored',
       'Highest_Score', 'Batting_Average', 'Balls_Faced',
       'Batting_Strike_Rate', 'Centuries', 'Half_Centuries', 'Fours', 'Sixes',
       'Catches_Taken', 'Stumpings', 'Balls_Bowled', 'Runs_Conceded',
       'Wickets_Taken', 'Best_Bowling_Match', 'Bowling_Average',
       'Economy_Rate', 'Bowling_Strike_Rate', 'Four_Wicket_Hauls',
       'Five_Wicket_Hauls']
```

```python
dp.Highest_Score
```

```python
dp['Highest_Score'] = dp['Highest_Score'].astype(str)
dp.Highest_Score = dp.Highest_Score.str.replace(r'[*]', '', regex=True)
```

```python
dp.Highest_Score # cleaned
```

```python
dp.Best_Bowling_Match
```

```python
dp['Best_Bowling_Match_2'] = dp['Best_Bowling_Match'].astype(str) + ['/0']
dp['Best_Bowling'] = dp['Best_Bowling_Match_2'].str.split('/')
dp.drop('Best_Bowling_Match_2', axis=1, inplace=True)
dp
```

```python
# Split each list into sublists of length 2
dp['Best_Bowling'] = dp['Best_Bowling'].apply(lambda x: [x[i:i+2] for i in range(0, len(x), 2)])
grouped_lists = dp.groupby('Player_Name')['Best_Bowling'].apply(lambda x: sum(x, []))
grouped_lists = grouped_lists.reset_index()
grouped_lists
```

```python
def clean_tuples(lst):
    new_lst = []
    for y in lst:
        try:
            # Ensure y is tuple/list of length 2
            if isinstance(y, (tuple, list)) and len(y) == 2:
                new_lst.append((int(y[0]), int(y[1])))
            else:
                # Fill missing elements with 0
                y0 = int(y[0]) if len(y) > 0 else 0
                y1 = int(y[1]) if len(y) > 1 else 0
                new_lst.append((y0, y1))
        except:
            # If y is not iterable at all, treat as (0,0)
            new_lst.append((0,0))
    return new_lst

grouped_lists['Best_Bowling'] = grouped_lists['Best_Bowling'].apply(clean_tuples)
```

```python
grouped_lists['Best_Bowling'] = grouped_lists['Best_Bowling'].apply(
    lambda lst: sorted(lst, key=lambda y: (-int(y[0]), int(y[1])))[0]
)
grouped_lists
```

```python
# Create new column as string with '/'
grouped_lists['Best_Bowling_Match'] = grouped_lists['Best_Bowling'].apply(lambda x: f"{x[0]}/{x[1]}")
grouped_lists
```

```python
dp.drop('Best_Bowling', axis=1, inplace=True)
```

```python
# Custom function to convert to numeric, and leave non-numeric columns as is
def convert_to_numeric(col):
    try:
        # Attempt to convert the column to numeric
        return pd.to_numeric(col, errors='ignore')  # Non-numeric values will become NaN
    except ValueError:
        # If it fails (e.g., non-numeric column), return the column as is
        return col

# Apply the function to all columns
dp = dp.apply(convert_to_numeric)
dp.dtypes
```

```python
agg_col = ['Matches', 'Not_Outs', 'Runs_Scored', 'Balls_Faced', 'Centuries', 'Half_Centuries', 'Fours', 'Sixes',
       'Catches_Taken', 'Stumpings', 'Balls_Bowled',
       'Runs_Conceded', 'Wickets_Taken', 'Four_Wicket_Hauls', 'Five_Wicket_Hauls']

dp_sum = dp[['Player_Name'] + agg_col].groupby('Player_Name').sum()
dp_sum.reset_index(inplace=True)
dp_sum
```

```python
df_players = pd.concat([dp_sum, grouped_lists.drop('Player_Name', axis=1)], axis=1)
df_players
```

```python
# Float aggregate - avg
# agg_col = ['Batting_Average', 'Batting_Strike_Rate', 'Bowling_Average', 'Economy_Rate', 'Bowling_Strike_Rate']
df_players['Batting_Average'] = df_players['Runs_Scored']/(df_players['Matches'] - df_players['Not_Outs'])
df_players['Batting_Strike_Rate'] = df_players['Runs_Scored']/df_players['Balls_Faced'] * 100
df_players['Bowling_Average'] = df_players['Runs_Conceded']/df_players['Wickets_Taken']
df_players['Economy_Rate'] = df_players['Runs_Conceded']/(df_players['Balls_Bowled']/6)
df_players['Bowling_Strike_Rate'] = df_players['Balls_Bowled']/df_players['Wickets_Taken']
df_players.replace([np.nan, np.inf, -np.inf], 0, inplace=True)
df_players
```
